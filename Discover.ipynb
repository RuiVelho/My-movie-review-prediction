```python
link2='https://datasets.imdbws.com/title.akas.tsv.gz'
title_df=pd.read_csv(link2, sep="\t", compression='gzip')
title_df
```

```python
link7='https://datasets.imdbws.com/title.ratings.tsv.gz'      #check
title_df7=pd.read_csv(link7, sep="\t", compression='gzip')
title_df7
```

```python
link3='https://datasets.imdbws.com/title.basics.tsv.gz'    #Check
title_df3=pd.read_csv(link3, sep="\t", compression='gzip')
title_df3=title_df3.merge(title_df7, how='inner', on='tconst')
title_df3
```

```python
#re.findall("\N" ,title_df3['startYear'])
title_df3['startYear'].replace(r'\N', 0, inplace=True)#.value_counts()/title_df3.shape[0]*100
#title_df3['startYear'].value_counts()

title_df3['startYear_int']=title_df3['startYear'].astype(int)
title_df3.info()
```

```python
title_df3=title_df3[title_df3['startYear_int']>=2000]
```

```python
title_df3=title_df3.loc[(title_df3['titleType']=='movie')]#.value_counts()
print(title_df3.shape)

title_df3['genres'].value_counts()/title_df3.shape[0]*100
```

```python
title_df3.drop(columns=['startYear', 'endYear'], inplace= True)
```

```python
title_df3.to_csv("titles_movies.csv")
```
